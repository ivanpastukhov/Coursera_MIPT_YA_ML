# Programming assignment #1

```python
f = open('cats_sent.txt', 'r')
```

```python
text = f.readlines()
f.close()
```

```python
t_split = map(lambda line: re.findall('[a-z]+', line.lower()), text)
```

```python
word_box = [word for line in t_split for word in line] #all words from the text
```

```python
word_count = {word:word_box.count(word) for word in word_box} #counted frequencies of words in the text
```

```python
word_index = {index:word for index, word in enumerate(word_count)} #indexed words
```

```python
word_vec   = np.array([[line.count(word) for word in word_index.itervalues()] for line in t_split])
print(word_vec[[1]])
```

```python
from scipy.spatial.distance import cosine
```

Find cosine distance between 1'st sentence and other ones:

$$
1 - \frac{u \cdot v}
            {||u||_2 ||v||_2}
            $$

```python
cos_dist_0_ = [cosine(word_vec[0], word_vec[index]) for index,value in enumerate(word_vec)]
```

```python
cos_dist_0_
```

```python
cos_dist_0_dict = {value:index for index,value in enumerate(cos_dist_0_)}
```

# Programming assignment #2

```python
from scipy.linalg import solve
import numpy as np
from numpy import sin, exp
import matplotlib.pyplot as plt
```

```python
def poly_coeff (X, order):
    return [[value ** i for i in range(order+1)] for index, value in enumerate(X)]
```

```python
x_boundary_1 = np.array([1.,15.])
x_boundary_2 = np.array([1.,8.,15.])
x_boundary_3 = np.array([1.,4.,10.,15.])

x1 = poly_coeff(x_boundary_1, 1)
x2 = poly_coeff(x_boundary_2, 2)
x3 = poly_coeff(x_boundary_3, 3)
x_linspace = np.linspace(1, 15, 50)
```

```python
basic_function = lambda x: sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
x = x_linspace
y = basic_function(x)
plt.plot(x,y)
plt.show()
```

## ________________________________________
## 1'st-order polynomial approximation:

```python
y1 = basic_function(x_boundary_1)
print('x1: ', x_boundary_1)
print('y1: ', y1)
b1 = solve(x1,y1)
print('b1: ', b1)
poly_1_function = (lambda x: b1[0] + b1[1]*x)
x1 = x_linspace
y1 = poly_1_function (x)
plt.plot(x,y,x1,y1)
plt.show()
```

## 2'nd-order polynomial approximation:

```python
y2 = basic_function(x_boundary_2)
b2 = solve(x2,y2)
print('x2: ', x2)
print('y2: ', y2)
print('b2: ', b2)
poly_2_function = (lambda x: b2[0] + b2[1]*x + b2[2]*x**2)
x2 = x_linspace
y2 = poly_2_function (x)
plt.plot(x,y,x1,y1,x2,y2)
plt.show()
```

```python
y3 = basic_function(x_boundary_3)
b3 = solve(x3,y3)
print('x3: ', x3)
print('y3: ', y3)
print('b3: ', b3)
poly_3_function = (lambda x: b3[0] + b3[1]*x + b3[2]*x**2+ b3[3]*x**3)
x3 = x_linspace
y3 = poly_3_function (x)
plt.plot(x,y,x1,y1,x2,y2,x3,y3)
plt.show()
```
