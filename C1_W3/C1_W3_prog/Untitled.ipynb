```python
from scipy.optimize import minimize, differential_evolution
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, exp
```

# Minimizing smooth function

```python
func = lambda x: sin(x / 5.) * exp(x / 10.) + 5. * exp(-x / 2.)
```

```python
x = np.linspace(1,30,100)
```

```python
y = func(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
min_1 = minimize(func, 2, method = 'BFGS')
x_min_1 = min_1.x
y_min_1 = func(x_min_1)

plt.plot(x,y,x_min_1, y_min_1,'ro')
plt.show()
```

```python
min_2 = minimize(func, 30, method = 'BFGS')
x_min_2 = min_2.x
y_min_2 = func(x_min_2)

plt.plot(x,y,x_min_1, y_min_1, 'ro', x_min_2, y_min_2,'go')
plt.show()
```

```python
print(zip(y_min_1, y_min_2))
```

# Global minimizing smooth function

```python
min_3 = differential_evolution(func, [(1.,30.)])
x_min_3 = min_3.x
y_min_3 = func(x_min_3)

plt.plot(x,y,x_min_3, y_min_3,'ro')
plt.show()
```

```python
print(y_min_3)
```

# Imperfect function minimizing

```python
func_dscr = lambda v: np.array(func(v)).astype(int)
y = func_dscr(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
#BFGS
min_4 = minimize(func_dscr, 30, method = 'BFGS')
x_min_4 = min_4.x
y_min_4 = func_dscr (x_min_4)
```

```python
#Diff. evolution
min_5 = differential_evolution(func_dscr, [(1,30)])
x_min_5 = min_5.x
y_min_5 = func_dscr (x_min_5)
```

```python
plt.plot(x,y,x_min_4,y_min_4, 'ro', x_min_5,y_min_5, 'gs')
plt.show()
```

```python
print(y_min_4, y_min_5)
```
