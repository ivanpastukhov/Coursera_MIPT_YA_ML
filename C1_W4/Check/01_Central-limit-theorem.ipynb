```python
from numpy import linspace
from numpy.random import choice
import matplotlib.pyplot as plt
from scipy.stats import beta, norm
%matplotlib inline

# we use this line for all plots to have the same size 
plt.rcParams['figure.figsize'] = (8, 5)
```

# Beta distribution
Quick reference from Wikipedia:
In probability theory and statistics, the beta distribution is a family of continuous probability distributions defined on the interval $[0, 1]$ parametrized by two positive shape parameters, denoted by $\alpha$ and $\beta$, that appear as exponents of the random variable and control the shape of the distribution.

We are going to work through this excercise using $\alpha=\beta=0.7,$ but you are welcome to change values of a and b below to see how this all works in other cases.

```python
# setting the parameters of distribution
a = 0.7
b = 0.7

# 'create' beta distrubution
beta_rv = beta(a, b)

# generate 1000 samples from a distribution
sample = beta_rv.rvs(1000)
```

```python
# plot our samples separated into 40 bins 
# you can also try changing 'bins' parameter here
plt.hist(sample, bins=40, normed=True, label='fraction of samples')

# also plot a theoretical pdf on the same plot
x = linspace(0,1,1000)
pdf = beta_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', color='red')

# all the following is finetuning
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
plt.close()
```

## Mean of beta distribution:
The expected value (mean) ($\mu$) of a Beta distribution random variable $X$ with two parameters $\alpha$ and $\beta$ is a function of only the ratio $\dfrac{\beta}{\alpha}$ of these parameters:

$$\mu = \operatorname{E}[X] = \int_0^1 x f(x;\alpha,\beta)\,dx = \int_0^1 x \,\frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}\,dx = \frac{\alpha}{\alpha + \beta} = \frac{1}{1 + \frac{\beta}{\alpha}}$$

Now, setting $\alpha=\beta=0.7$ gives $\mu = 0.5.$

```python
# we do not hardcode a value of mu so that it will change if you alter a and b at the beginning of the excercise 
mu = 1 / (1 + b / a)
```

## Variance of beta distribution:
The variance (the second moment centered on the mean) of a Beta distribution random variable $X$ with parameters $\alpha$ and $\beta$ is:

$$\sigma^2=\operatorname {E} \left[(X-\mu )^{2}\right]=\frac{\alpha \beta }{(\alpha +\beta)^{2}(\alpha +\beta +1)}$$

Now, setting $\alpha=\beta=0.5$ gives $\sigma^2 = \dfrac{5}{48}.$

```python
sigma2 = (a * b) / ((a + b)**2 * (a + b + 1))
```

## CLT:
Now, remember that CLT says that mean value of an approximation will be the same and variance will be $n$ times smaller, where $n$ is a size of a batch/sample. This gives us the following expressions for $\mu_n$ and $\sigma_n:$ 
$$\mu_n = \mu, \qquad \sigma_n^2 = \dfrac{\sigma^2}{n}$$
Also notice that we will pass standard deviation to the generator of normal distribution, not variance. As standard deviation is the square root of variance, we obtain the following expression:
$$\text{std}_n = \sigma_n = \sqrt{\sigma_n^2} = \sqrt{\dfrac{\sigma^2}{n}}$$

```python
# we use a list here to avoid code duplication
N = [3, 10, 30]
for n in N:
    
    # note that we have to pass std = sqrt(sigma^2) to the generator of normal distribution
    aprox_norm_dist = norm(mu, (sigma2 / n)**(1 / 2))
    
    # generate batches each of n randomly chosen variables from sample
    # change this value to see how the quality of approximation depends on the numbers of batches
    batches = 1000 
    means = []
    for i in range(batches):
        X = beta_rv.rvs(n)
        
        # function sum() returns a sum of all elements of a list
        mean = sum(X) / n
        means.append(mean)
    
    # plot histogram of out 10000 means separated into 40 bins
    # note that it is necessary to pass 'range' parameter or else it will take [x_min, x_max]
    # in other words, it will shrink the interval so that you won't see how the variance decreases
    plt.hist(means, bins=40, normed=True, range=[0,1], label='sample means, $n=$'+str(n))

    # plot theoretical pdf of normal distribution which approximates our dist
    x = linspace(0,1,1000)
    pdf = aprox_norm_dist.pdf(x)
    
    # note that we use 'color' parameter to make our graphs distinguishable
    plt.plot(x, pdf, label='approximate pdf', alpha=1, color='r')
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.legend()
    
    # we use 'show' and 'close' methods to separate our data into len(N) different plots
    # you can cut these to see what will happen
    plt.show()
    plt.close()
```

## Conclusion
As we can see, the quality of approximation grows with $n$ increasing. 
It will also increase as number of batches grows.
