Импортируем необходимые библиотеки

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Сгенерируем выборку объема 1000 для [гамма-распределения](https://ru.wikipedia.org/wiki/%D0%93%D0%B0%D0%BC%D0%BC%D0%B0-%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5)  с параметрами k = 9 и THETA = 0.5. Нарисуем на одном графике гистограмму выборки и плотность функции гамма-распределения.


```python
rv = sts.gamma(9, scale=0.5)
sample = rv.rvs(1000)
 
x = np.linspace(rv.ppf(0.001),rv.ppf(0.999),1000)
pdf = rv.pdf(x)
 
plt.plot(x, pdf, label='theoretical pdf',)
plt.hist(sample,normed=True,color='grey',label='actual')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Используя теоретические сведения, посчитаем мат ожидание и среднеквадратичное отклонение нашего гамма-распределения (9, 0.5)

```python
K = 9
THETA = 0.5
MU = K * THETA
DIS = K * THETA**2
SIGMA = DIS ** 0.5
print 'Мат ожидание равно =', MU
print 'Квадратное отклонение равно =', SIGMA
```

Для оценки распределения выборочного среднего объма n, определим специальную функцию clt (Central limit theorem).

```python
def clt(n, sub_quantity=1000):
    means = []
    for i in range(sub_quantity):
        means.append(np.mean(np.random.gamma(9, scale=0.5, size=n)))
    
    #Параметры данного нормального распределения подсчитаны согласно ЦПТ
    norm_rv = sts.norm(MU, SIGMA/n**0.5)
    x = np.linspace(norm_rv.ppf(0.001),norm_rv.ppf(0.999),100)
    
    pdf = norm_rv.pdf(x)
    print 'Распределение выборочных средних (n = ' + repr(n) + ')'
    plt.plot(x, pdf, label='theoretical CDF')
    plt.hist(means, normed=True, label='actual')
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

```python
clt(5)
```

```python
clt(10)
```

```python
clt(50)
```
