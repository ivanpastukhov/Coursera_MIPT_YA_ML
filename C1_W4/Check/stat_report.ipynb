```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

# Центральная предельная теорема
## Экспоненциальное распределение
Случайная величина распределена по экспоненциальному законц с параметром $\lambda > 0$, если плотность её распределения представляется в виде:

\begin{equation}
f(x) = 
\begin{cases}
\lambda e^{-\lambda x}, & x \geqslant 0 \\
0, & x < 0
\end{cases}
\end{equation}

Сгенерируем выборку размером $1000$ из экспоненциального распределения с параметром $\lambda=4$ ($1/\lambda=0.25$):

```python
expon_rv = sts.expon(0, 0.25)
sample = expon_rv.rvs(1000)
```

Построим функцию распределения экспоненциальной случайной велечины и гистограмму по выборке:

```python
x = np.linspace(0,3,31)

plt.plot(x, expon_rv.pdf(x), label='PDF')
plt.hist(sample, normed=True, label='Histogram')

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend()
```

## Оценка распределения выборочного среднего

Создание выборок и построение гистограмм с нанесёнными графиками нормальных распределений удобнее выполнять вместе в одном цикле, поэтому сначала вычислим параметры нормальных распределений для приближения распределения выборочных средних.
Итак, пусть, как и указано в задании, размеры выборок будут $n_1=5$, $n_2=10$ и $n_3=50$:

```python
n = [5, 10, 50]
```

Пусть $X$ -- случайная величина, распределённая по экспоненциальному закону. Тогда
$$ \mathbb{E}[X] = \frac{1}{\lambda} = 0.25, \quad D[X] = \frac{1}{\lambda^2} = 0.0625 $$

Параметры нормальных распределений (математическое ожидание $\mu$ и среднеквадратическое отклонение $\sigma$) согласно ЦПТ задаются следующим образом:
\begin{eqnarray}
    \mu_i &=& \mathbb{E}[X] = \frac{1}{\lambda} = 0.25, \quad i = 1,2,3 \\
    \sigma_i &=& \sqrt{\frac{D[X]}{n_i}} = \sqrt{\frac{1}{n_i \lambda^2}} = \frac{1}{4\sqrt{n_i}}, \quad i = 1,2,3
\end{eqnarray}

```python
mu = expon_rv.mean()
sigma = np.zeros(3)
for i in range(3):
    sigma[i] = expon_rv.std() / (n[i] ** 0.5)
print mu, sigma
```

Теперь строим гистограммы и наносим графики нормальных распределений, оценивающих распределение выборочных средних:

```python
smpl_num = 1000 # количество выборок
x = np.linspace(0,1,1000)

# создаём figure, на который выведем все три графика одновременно
plt.figure(figsize=(18, 5))

# для каждого размера выборки
for i in range(3):
    # составляем массив выборочных средних
    means = np.zeros(smpl_num)
    for j in range(smpl_num):
        smpl = expon_rv.rvs(n[i])
        means[j] = np.mean(smpl)
    
    # рисуем гистограмму
    plt.subplot(1,3,i+1)
    plt.hist(means, normed = True, label='Histogram')
    
    # рисуем график нормального распределения
    plt.plot(x, sts.norm(mu, sigma[i]).pdf(x), label='Normal distr. from CLT')
    
    # выводим необхожимые подписи
    plt.title('n = ' + str(n[i]))
    plt.xlabel('x'), plt.ylabel('fraction of samples')
    plt.xlim(0, 1)
    plt.ylim(0, 12)
    plt.legend()

plt.show()
```

## Вывод

С увеличением количества элементов в выборке распределение становится более сфокусированным: разброс сокращается, а график "вытягивается" вверх. Помимо этого гистограмма всё больше становится похожа на гистограмму нормального распределения, что согласуется с ЦПТ, гласящей (нестрого): чем больше n, тем скорее распределение выборочных средних будет похоже на нормальное.
